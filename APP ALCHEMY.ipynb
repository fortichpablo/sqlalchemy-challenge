```python
# Import necessary libraries
import pandas as pd
import numpy as np
import datetime as dt
import sqlalchemy 
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

import flask
from flask import Flask, jsonify

# Define the file path for the SQLite database
database_path = "/Users/pablofortich/Desktop/SQL ALCHEMY CHALLENGE/Resources/hawaii.sqlite"

# Create an engine to connect to the SQLite database
engine = create_engine(f"sqlite:///{database_path}")

# Use SQLAlchemy to reflect the database tables
Base = automap_base()
Base.prepare(engine, reflect=True)

# Create references to the Measurement and Station tables
Measurement = Base.classes.measurement
Station = Base.classes.station

# Create a session to interact with the database
session = Session(engine)

# Create a Flask app instance
app = Flask(__name__)

# Define a route to retrieve precipitation data
@app.route("/api/v1.0/precipitation")
def precipitation():
    # Query precipitation and date data from the Measurement table
    m_prcp = session.query(Measurement.prcp, Measurement.date).\
        filter(Measurement.date > '2016-08-23').\
        order_by(Measurement.date).all()
    
    # Convert the query results into a dictionary
    prdict = {date: x for date, x in m_prcp}
    
    # Return the dictionary as JSON
    return jsonify(prdict)

# Define the root route to display available routes
@app.route("/")
def welcome():
    return (f"Welcome page <br/>"
            f"Routes: <br/>"
            f"/api/v1.0/precipitation <br/>"
            f"/api/v1.0/stations <br/>"
            f"/api/v1.0/tobs <br/>")

# Define a route to retrieve a list of weather stations
@app.route("/api/v1.0/stations")
def station():
    # Query station data from the Station table
    result = session.query(Station.station).all()
    
    # Convert the query results into a list
    st_list = list(np.ravel(result))
    
    # Return the list as JSON
    return jsonify(st_list)

# Define a route to retrieve temperature observations
@app.route("/api/v1.0/tobs")
def tobs():
    # Query temperature observations for a specific station (USC00519281)
    tobss = session.query(Measurement.tobs).\
            filter(Measurement.station == 'USC00519281').\
            filter(Measurement.date >= '2017,8,23').all()
    
    # Convert the query results into a list
    tobs_list = list(np.ravel(tobss))
    
    # Return the list as JSON
    return jsonify(tobs_list)

# Define a route to retrieve temperature statistics for a given date range
@app.route("/api/v1.0/<start>/<end>")
def temps(start, end):
    # Query temperature data for a given date range
    findings = session.query(Measurement).filter(Measurement.date >= start).filter(Measurement.date <= end)
    
    # Extract temperature values and store in a list
    found = [] 
    for row in findings:
        found.append(row.tobs) 
    
    # Return temperature statistics as JSON
    return jsonify({"tempmin": min(found), "tempmax": max(found), "tempavg": np.mean})

# Run the Flask app
if __name__ == "__main__":
    app.run(debug=True)


```
